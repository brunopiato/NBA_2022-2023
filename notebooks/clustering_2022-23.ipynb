```python
# Importing the complete dataset for all seasons
path = '/home/bruno/repos/NBA_2022-2023/data/scraped_all_seasons'

df_bios_complete = pd.read_csv(path+'/df_bios_complete.csv', low_memory=False)
df_clutch_advanced_complete = pd.read_csv(path+'/df_clutch_advanced_complete.csv', low_memory=False)
df_clutch_misc_complete = pd.read_csv(path+'/df_clutch_misc_complete.csv', low_memory=False)
df_clutch_scoring_complete = pd.read_csv(path+'/df_clutch_scoring_complete.csv', low_memory=False)
df_clutch_traditional_complete = pd.read_csv(path+'/df_clutch_traditional_complete.csv', low_memory=False)
df_clutch_usage_complete = pd.read_csv(path+'/df_clutch_usage_complete.csv', low_memory=False)
# df_defense_2P = pd.read_csv(path+'/df_defense_2P.csv', low_memory=False) # Beginning in the season 2013-14
# df_defense_3P = pd.read_csv(path+'/df_defense_3P.csv', low_memory=False) # Beginning in the season 2013-14
# df_defense_6ft = pd.read_csv(path+'/df_defense_6ft.csv', low_memory=False) # Beginning in the season 2013-14
# df_defense_10ft = pd.read_csv(path+'/df_defense_10ft.csv', low_memory=False) # Beginning in the season 2013-14
# df_defense_15ft = pd.read_csv(path+'/df_defense_15ft.csv', low_memory=False) # Beginning in the season 2013-14
# df_defense_overall = pd.read_csv(path+'/df_defense_overall.csv', low_memory=False)
df_hustle_complete = pd.read_csv(path+'/df_hustle_complete.csv', low_memory=False)
df_players_advanced_complete = pd.read_csv(path+'/df_players_advanced_complete.csv', low_memory=False)
df_boxout_complete = pd.read_csv(path+'/df_players_boxout_complete.csv', low_memory=False)
df_defense_complete = pd.read_csv(path+'/df_players_defense_complete.csv', low_memory=False)
df_scoring_complete = pd.read_csv(path+'/df_players_scoring_complete.csv', low_memory=False)
df_shoting_byzone = pd.read_csv(path+'/df_shooting_byzone.csv', low_memory=False)
```

## Combining all the datasets

```python
df = pd.merge(df_bios_complete, df_clutch_advanced_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_clutch_misc_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_clutch_scoring_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_clutch_traditional_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_clutch_usage_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_hustle_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_players_advanced_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_boxout_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_defense_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_scoring_complete, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df = pd.merge(df, df_shoting_byzone, 
              how = 'left', 
              on = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ABBREVIATION', 'TEAM_ID', 'season_id', 'season_type'],
              suffixes = ['', '_DROP'])

df.drop(df.filter(regex='_DROP$').columns, axis=1, inplace=True)
```

```python
df.drop(df.filter(regex='_RANK$').columns, axis=1, inplace=True)
```
